```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Base customers and products
customers = [
    ("C001", "Aarav"), ("C002", "Meera"), ("C003", "Rahul"),
    ("C004", "Ananya"), ("C005", "Vikram"), ("C006", "Priya"),
    ("C007", "Rohan"), ("C008", "Neha"), ("C009", "Karan"), ("C010", "Isha")
]

regions = ["North", "South", "East", "West"]

products = [
    ("Technology", "Laptop", 75000),
    ("Technology", "Tablet", 30000),
    ("Technology", "Monitor", 22000),
    ("Furniture", "Office Chair", 12000),
    ("Furniture", "Desk", 25000),
    ("Office Supplies", "Notebook", 800),
    ("Office Supplies", "Pen Set", 1200)
]

rows = []
start_date = datetime(2023, 1, 1)

for i in range(30000):
    order_id = 2000 + i
    order_date = start_date + timedelta(days=random.randint(0, 365))
    customer_id, customer_name = random.choice(customers)
    region = random.choice(regions)
    category, product, base_price = random.choice(products)
    quantity = random.randint(1, 5)
    sales = base_price * quantity

    rows.append([
        order_id,
        order_date.strftime("%Y-%m-%d"),
        customer_id,
        customer_name,
        region,
        category,
        product,
        sales,
        quantity
    ])

df = pd.DataFrame(rows, columns=[
    "order_id", "order_date", "customer_id", "customer_name",
    "region", "category", "product_name", "sales", "quantity"
])

df.to_csv("../data/orders_large.csv", index=False)

print("orders_large.csv generated with", len(df), "rows")
```
